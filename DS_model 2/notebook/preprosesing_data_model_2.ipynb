```python
import pandas as pd
import re

# 1. Memuat dataset yang kotor
df_dirty = pd.read_csv('/content/data_sintesis_wellbeing_kotor.csv')
print("Ukuran Awal Data Kotor:", df_dirty.shape)

# ==========================================
# PROSES PREPROCESSING (PEMBERSIHAN DATA)
# ==========================================

# Langkah 1: Standarisasi dan Pemetaan Kembali Nama Kolom
clean_col_mapping = {
    ' usia_anda': 'Usia Anda',
    'jenis_kelamin ': 'Jenis Kelamin',
    'jam_tidur_semalam': 'Jam Tidur Semalam',
    'seberapa_sibuk_anda_hari_ini_(1-5)': 'Seberapa Sibuk Anda Hari Ini (1-5)',
    'seberapa_banyak_tugas_pekerjaan_(1-5)': 'Seberapa Banyak Tugas/Pekerjaan (1-5)',
    'seberapa_tertekan_anda_(1-5)': 'Seberapa Tertekan Anda (1-5)',
    'suasana_hati_anda_(1-5)': 'Suasana Hati Anda (1-5)',
    'seberapa_sering_merasa_lelah_emosional_(1-5)': 'Seberapa Sering Merasa Lelah Emosional (1-5)',
    'seberapa_termotivasi_anda_(1-5)': 'Seberapa Termotivasi Anda (1-5)',
    'ceritakan_hari_anda_hari_ini ': 'Ceritakan Hari Anda Hari Ini'
}
df_dirty.rename(columns=clean_col_mapping, inplace=True)

# Langkah 2: Membersihkan Kolom 'Usia Anda'
# Menghapus spasi luar, mengubah ke huruf kecil, lalu memetakan ke kategori asli
df_dirty['Usia Anda'] = df_dirty['Usia Anda'].str.strip().str.lower()
usia_mapping = {
    '21-23 tahun': '21 - 23 tahun',
    '18-20 tahun': '18 - 20 tahun',
    '24-26 tahun': '24 - 26 tahun',
    '>26 tahun': '> 26 tahun',
    '18 tahun': '18 tahun'
}
df_dirty['Usia Anda'] = df_dirty['Usia Anda'].map(usia_mapping)

# Langkah 3: Membersihkan Kolom 'Jenis Kelamin'
# Menghapus spasi dan menstandarkan huruf kapital di awal kata (Capitalize)
df_dirty['Jenis Kelamin'] = df_dirty['Jenis Kelamin'].str.strip().str.lower().str.capitalize()

# Langkah 4: Membersihkan Kolom 'Jam Tidur Semalam'
# Mengekstrak hanya karakter angka menggunakan Regular Expression (Regex) lalu konversi ke Integer
df_dirty['Jam Tidur Semalam'] = df_dirty['Jam Tidur Semalam'].astype(str).str.extract(r'(\d+)').astype(int)

# Langkah 5: Membersihkan Kolom Numerik Skala 1-5
# Menghapus spasi, menghilangkan format desimal .0 jika ada, lalu konversi ke Integer
target_numeric_cols = [
    'Seberapa Sibuk Anda Hari Ini (1-5)', 'Seberapa Banyak Tugas/Pekerjaan (1-5)',
    'Seberapa Tertekan Anda (1-5)', 'Suasana Hati Anda (1-5)',
    'Seberapa Sering Merasa Lelah Emosional (1-5)', 'Seberapa Termotivasi Anda (1-5)'
]
for col in target_numeric_cols:
    df_dirty[col] = df_dirty[col].astype(str).str.strip().str.replace(r'\.0$', '', regex=True).astype(int)

# Langkah 6: Membersihkan Kolom Cerita/Teks Panjang
# Menghapus spasi kosong yang tidak perlu di awal dan akhir kalimat
df_dirty['Ceritakan Hari Anda Hari Ini'] = df_dirty['Ceritakan Hari Anda Hari Ini'].str.strip()

nama_file_output = 'data_sintesis_wellbeing_clean.csv'
df_dirty.to_csv(nama_file_output, index=False)

print(f"Data yang sudah bersih berhasil disimpan dengan nama: '{nama_file_output}'")
```
